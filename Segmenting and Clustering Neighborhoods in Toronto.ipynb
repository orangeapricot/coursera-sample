# Segmenting and Clustering Neighborhoods in Toronto

```python
#install pacages
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library
from bs4 import BeautifulSoup as bs
```

## Web Scraping

```python
webpage = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
page = bs(webpage.content,'html.parser')
```

```python
data= str(page.table)
```

```python
dfs = pd.read_html(data)
df=dfs[0]
df.head()
```

```python
# Ignore cells with a borough that is Not assigned.
df1 = df[df.Borough != 'Not assigned']
# dealing with the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# dealing with the cell has a borough but a Not assigned neighborhood
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])

df2
```

```python
df2.shape
```

## Get the latitudes and longitudes

```python
lat_data = pd.read_csv('https://cocl.us/Geospatial_data')
lat_data.head()
```

```python
# merge the dataframe 
df_lat = pd.merge(df2,lat_data,on='Postal Code')
df_lat.head()
```

## Cluster

```python
#the rows which contains Toronto in their Borough.
df_Toronto = df_lat[df_lat['Borough'].str.contains('Toronto',regex=False)]
df_Toronto
```

```python
# visualization
map_toronto = folium.Map(location=[43.651070,-79.347015], zoom_start=11)

for lat, lng, borough,neighbourhood in zip(df_Toronto['Latitude'],df_Toronto['Longitude'],df_Toronto['Borough'],df_Toronto['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
# k-means cluster
k=5
toronto_clustering = df_Toronto.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
print(kmeans.labels_[0:10])
df_Toronto['Labels']=kmeans.labels_
df_Toronto
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015], zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_Toronto['Latitude'], df_Toronto['Longitude'], df_Toronto['Neighborhood'], df_Toronto['Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
